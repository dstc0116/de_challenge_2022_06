```python
import pandas as pd
import numpy as np
```

read data

```python
df = pd.read_csv('car/car.data',names= ['buying','maint', 'doors', 'persons', 'lug_boot', 'safety','classes'])
df
```

Value mapping

```python
mapping ={'vhigh':4,'high':3,'med':2,'low':1,
        '5more':6,'more':5,
        'small' :1, 'med':2, 'big':3,
        'unacc':1, 'acc':2,'good':3, 'vgood':4
        }
df_map = df.replace(mapping)
df_map
```

Model training 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics  import f1_score,accuracy_score,classification_report

#  convert from pandas to numpy
car = df_map.values

# prepare dataset
X,y = car[:,1:], car[:,1]
X,y = X.astype(int), y.astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)

model = RandomForestClassifier(max_depth=None,
                              max_features='auto',
                              n_estimators=500,
                              random_state=0)
model.fit(X_train,y_train)

# We use the predict() on the model to predict the output
pred=model.predict(X_test)

print(accuracy_score(y_test,pred))
target_names = ['vhigh','high','med','low']
print(classification_report(y_test, pred, target_names=target_names))
```
